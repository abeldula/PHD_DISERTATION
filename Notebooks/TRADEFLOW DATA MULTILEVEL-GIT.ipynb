```python
def findCountryAlpha2(country_name):
    try:
        return pycountry.countries.get(name=country_name).alpha_2
    except:
        return ("not founded")
def findCountryAlpha3(country_name):
    try:
        return pycountry.countries.get(name=country_name).alpha_3
    except:
        return ("not founded")
def findCountryNumeric(country_name):
    try:
        return pycountry.countries.get(name=country_name).numeric
    except:
        return ("not founded")
def findCountryOfficialName(country_name):
    try:
        return pycountry.countries.get(name=country_name).official_name
    except:
        return ("not founded")
def findCountryAlpha31(country_name):
    try:
        return pycountry.countries.get(alpha_3=country_name).official_name
    except:
        return ("not founded")
def findCountryAlpha32(country_name):
    try:
        return pycountry.countries.get(alpha_3=country_name).numeric
    except:
        return ("not founded")
```

### PROCESSING THE DATA FOR EXPORTING IT AS CLEAN

The Grubel–Lloyd index measures intra-industry trade of a particular product. It was introduced by Herb Grubel and Peter Lloyd in 1971.
\begin{equation*}
\frac{(X_i + M_i)-|X_i - M_i|}{X_i + M_i} = 1- \frac{|X_i - M_i|}{X_i + M_i} ; 0\leq GL_i \leq 1
\end{equation*}
where Xi denotes the export, Mi the import of good i.
If GLi = 1, there is a good level of intra-industry trade. This means for example the Country in consideration Exports the same quantity of good i as much as it Imports. Conversely, if GLi = 0, there is no intra-industry trade at all. This would mean that the Country in consideration only either Exports or only Imports good i. <br>
<b>uv -- shows value of exports in (billions USD) or value of imports in (billions USD)<b/><br>
<b>v -- shows volume of exports or volume of imports<b/><br>
<b>k -- shows HSCODE of exported OR imported Goods<b/>

```python
# Perform PooledOLS
from linearmodels import PooledOLS
from linearmodels import PanelOLS
import statsmodels.api as sm
mod = PooledOLS(MERGED3.shdi, MERGED3['lgnic'])#, entity_effect=True
#mod = PanelOLS(MERGED3.shdi, MERGED3[['lgnic','Mining,']], entity_effect=True)
# Store values for checking homoskedasticity graphically
pooledOLS_res = mod.fit(cov_type='clustered', cluster_entity=True)
fittedvals_pooled_OLS = pooledOLS_res.predict().fitted_values
residuals_pooled_OLS = pooledOLS_res.resids

#from linearmodels import PanelOLS
#mod = PanelOLS(data.invest, data[['value','capital']], entity_effect=True)
#res = mod.fit(cov_type='clustered', cluster_entity=True)
```

```python
# 3A. Homoskedasticity
#how to identify heteroscedasticity graphically and White-Test and Breusch-Pagan-Test (similar)
import matplotlib.pyplot as plt
 # 3A.1 Residuals-Plot for growing Variance Detection
fig, ax = plt.subplots()
ax.scatter(fittedvals_pooled_OLS, residuals_pooled_OLS, color = "blue")
ax.axhline(0, color = 'r', ls = '--')
ax.set_xlabel("Predicted Values", fontsize = 15)
ax.set_ylabel("Residuals", fontsize = 15)
ax.set_title("Homoskedasticity Test", fontsize = 30)
plt.show()
```

Basically, a residuals-plot represents predicted values (x-axis) vs. residuals (y-axis). If the plotted data points spread out, this is an indicator for growing variance and thus, for heteroskedasticity. Since this seems to be the case in our example, we might have the first violation. But let´s check this with the White- and the Breusch-Pagan-Test:

```python
 #3A.2 White-Test
from statsmodels.stats.diagnostic import het_white, het_breuschpagan
pooled_OLS_dataset = pd.concat([MERGED3, residuals_pooled_OLS], axis=1)
pooled_OLS_dataset = pooled_OLS_dataset.drop(["Year"], axis = 1).fillna(0)
exog = sm.tools.tools.add_constant(MERGED3['lgnic']).fillna(0)
white_test_results = het_white(pooled_OLS_dataset["residual"], exog)
labels = ["LM-Stat", "LM p-val", "F-Stat", "F p-val"] 
print(dict(zip(labels, white_test_results)))
# 3A.3 Breusch-Pagan-Test
breusch_pagan_test_results = het_breuschpagan(pooled_OLS_dataset["residual"], exog)
print(dict(zip(labels, breusch_pagan_test_results)))
```

In simple terms, if p < 0.05, then heteroskedasticity is indicated. Both tests give very small p-values 
Therefore, we have proven our first violation! Let´s perform assumption 3b:

```python
# 3.B Non-Autocorrelation
# Durbin-Watson-Test
from statsmodels.stats.stattools import durbin_watson
durbin_watson_test_results = durbin_watson(pooled_OLS_dataset["residual"]) 
print(durbin_watson_test_results)
```

The Durbin-Watson-Test will have one output between 0 – 4. The mean (= 2) would indicate that there is no autocorrelation identified, 0 – 2 means positive autocorrelation (the nearer to zero the higher the correlation), and 2 – 4 means negative autocorrelation (the nearer to four the higher the correlation). In our example, the result is 0.16471149722210698, which clearly indicates strong positive autocorrelation.

As a consequence, assumption 3b is also violated, so it seems that a FE-/RE-model will be more suitable.

```python
# FE und RE model
from linearmodels import PanelOLS
from linearmodels import RandomEffects
exog = sm.tools.tools.add_constant(MERGED3['lgnic'])
endog = MERGED3.shdi
# random effects model
model_re = RandomEffects(endog, exog) 
re_res = model_re.fit() 
# fixed effects model
model_fe = PanelOLS(endog, exog, entity_effects = True) 
fe_res = model_fe.fit() 
#print results
print(re_res)
print(fe_res)
```

```python
import numpy.linalg as la
from scipy import stats
import numpy as np
def hausman(fe, re):
    b = fe.params
    B = re.params
    v_b = fe.cov
    v_B = re.cov
    df = b[np.abs(b) < 1e8].size
    chi2 = np.dot((b-B).T, la.inv(v_b-v_B).dot(b-B)) 
    pval = stats.chi2.sf(chi2, df)
    return chi2, df, pval
hausman_results = hausman(fe_res, re_res) 
print('chi-Squared: ' + str(hausman_results[0]))
print('degrees of freedom: ' + str(hausman_results[1]))
print('p-Value: ' + str(hausman_results[2]))
```

Since the p-value is very small (0.00015139558109921204), the null hypothesis can be rejected. Accordingly, the FE-model seems to be the most suitable, because we clearly have endogeneity in our model.

In order to model endogeneity, we could now perform regression models like 2SLS (2 Stage Least Squares) in which instrument variables help to deal with endogeneity

```python
TRADEFLOW_FULL_DATASET_PROCESS = TRADEFLOW_FULL_DATASET.groupby(
    ['t','iso3_i','iso3_j','bec_Processing'],).agg(
    trade_type_max=pd.NamedAgg(column="trade_type", aggfunc="max"),
    GL_mean=pd.NamedAgg(column="GL", aggfunc="mean"),
    k_count=pd.NamedAgg(column="k", aggfunc="count"),
    uv_sum=pd.NamedAgg(column="uv", aggfunc="sum"),
#    uv_mean=pd.NamedAgg(column="uv", aggfunc="mean"),
    v_sum=pd.NamedAgg(column="v", aggfunc="sum"),
#    v_mean=pd.NamedAgg(column="v", aggfunc="mean"),
    v_LATi=pd.NamedAgg(column="LATi", aggfunc="first"),#"nth(0)", lambda x: x.max()
    v_LONGi=pd.NamedAgg(column="LONGi", aggfunc="first"),
    v_LATj=pd.NamedAgg(column="LATj", aggfunc="first"),
    v_LONGj=pd.NamedAgg(column="LONGj", aggfunc="first"),    
    i=pd.NamedAgg(column="i", aggfunc="first"),
    j=pd.NamedAgg(column="j", aggfunc="first"),
#    v_continentj=pd.NamedAgg(column="continent_j", aggfunc="first"),
#    v_sub_continentj=pd.NamedAgg(column="sub_continent_j", aggfunc="first"),
#    v_bec_EndUse=pd.NamedAgg(column="bec_Category", aggfunc="first"),
#    v_BEC5Processing=pd.NamedAgg(column="bec_Processing", aggfunc="first"),
#    v_k2=pd.NamedAgg(column="k2", aggfunc="first")
    ).reset_index()
```

```python
TOPEXP_PARTNER_BYVAL = TRADEFLOW_FULL_DATASET_PROCESS.groupby(['t','iso3_j','bec_Processing']).apply(
    lambda x: x.nlargest(3, ['uv_sum'])).reset_index(drop=True).groupby(
    ['t','iso3_j','bec_Processing'])['iso3_i'].apply(list).reset_index().rename(
    columns={"iso3_i": "TOPEXP_PARTNER_BYVAL"})
TOPEXP_PARTNER_BYHS = TRADEFLOW_FULL_DATASET_PROCESS.groupby(['t','iso3_j','bec_Processing']).apply(
    lambda x: x.nlargest(3, ['k_count'])).reset_index(drop=True).groupby(
    ['t','iso3_j','bec_Processing'])['iso3_i'].apply(list).reset_index().rename(
    columns={"iso3_i": "TOPEXP_PARTNER_BYHS"})
TOPIMP_PARTNER_BYVAL = TRADEFLOW_FULL_DATASET_PROCESS.groupby(['t','iso3_i','bec_Processing']).apply(
    lambda x: x.nlargest(3, ['uv_sum'])).reset_index(drop=True).groupby(
    ['t','iso3_i','bec_Processing'])['iso3_j'].apply(list).reset_index().rename(
    columns={"iso3_j": "TOPIMP_PARTNER_BYVAL"})
TOPIMP_PARTNER_BYHS = TRADEFLOW_FULL_DATASET_PROCESS.groupby(['t','iso3_i','bec_Processing']).apply(
    lambda x: x.nlargest(3, ['k_count'])).reset_index(drop=True).groupby(
    ['t','iso3_i','bec_Processing'])['iso3_j'].apply(list).reset_index().rename(
    columns={"iso3_j": "TOPIMP_PARTNER_BYHS"})
```
